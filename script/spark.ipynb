# Task
Summarize the top 10 movies with the highest average ratings, considering only movies with at least 100 ratings, using the "movies.csv" and "ratings.csv" files.

## Calculate Rating Sum and Count

### Subtask:
Calculate the sum of ratings and the count of ratings for each movie ID from the 'rating_by_movieid' RDD.


**Reasoning**:
First, I'll transform the 'rating_by_movieid' RDD to include a count of 1 for each rating, creating pairs of `(movie_id, (rating, 1))`. Then, I'll use `reduceByKey` to aggregate these pairs by movie ID, summing the ratings and the counts to get `(movie_id, (sum_of_ratings, count_of_ratings))` as instructed.



```python
rating_sums_counts = rating_by_movieid.map(lambda x: (x[0], (x[1], 1))).reduceByKey(lambda val1, val2: (val1[0] + val2[0], val1[1] + val2[1]))
rating_sums_counts.take(5)
```

## Filter Movies by Minimum Ratings

### Subtask:
Filter out movies that have received less than 100 ratings from the `rating_sums_counts` RDD.


**Reasoning**:
I need to filter the `rating_sums_counts` RDD to include only movies with at least 100 ratings, as specified in the instructions. The rating count is the second element of the value tuple in the RDD.



```python
filtered_ratings_counts = rating_sums_counts.filter(lambda x: x[1][1] >= 100)
filtered_ratings_counts.take(5)
```

## Calculate Average Rating

### Subtask:
Calculate the average rating for each movie from the filtered data.


**Reasoning**:
To calculate the average rating for each movie, I will apply a map transformation to the `filtered_ratings_counts` RDD. This transformation will divide the sum of ratings by the count of ratings for each movie, storing the result as (movie_id, average_rating) in the `average_ratings` RDD. I will then display the first few results to verify.



```python
average_ratings = filtered_ratings_counts.map(lambda x: (x[0], x[1][0] / x[1][1]))
average_ratings.take(5)
```

## Join with Movie Titles

### Subtask:
Join the RDD of movies with their average ratings with the 'movies_by_movieid' RDD to get the movie titles.


**Reasoning**:
To combine the movie titles with their calculated average ratings, I will perform an inner join between the 'movies_by_movieid' RDD (containing movie IDs and titles) and the 'average_ratings' RDD (containing movie IDs and average ratings). This will yield an RDD with the structure (movie_id, (title, average_rating)), as specified in the instructions. I will then display the first 5 elements to verify the join.



```python
movie_titles_ratings = movies_by_movieid.join(average_ratings)
movie_titles_ratings.take(5)
```

## Sort and Select Top 10 Movies

### Subtask:
Sort the `movie_titles_ratings` RDD by average rating in descending order and select the top 10 movies.

**Reasoning**:
To sort the RDD by average rating in descending order, I need to extract the average rating from the tuple and use it as the sorting key. Then, I will use the `take` action to get the top 10 movies.



```python
top_10_movies = movie_titles_ratings.sortBy(lambda x: x[1][1], ascending=False).take(10)
for movie in top_10_movies:
    print(movie)
```

```python
movies_df = spark.read.csv(
    "/content/sample_data/movies.csv",
    header=True,
    inferSchema=True
)

ratings_df = spark.read.csv(
    "/content/sample_data/ratings.csv",
    header=True,
    inferSchema=True
)
```

```python
from pyspark.sql.functions import avg, count

ratings_agg = ratings_df.groupBy("movieId") \
    .agg(
        avg("rating").alias("avg_rating"),
        count("rating").alias("rating_count")
    )
```

```python
ratings_filtered = ratings_agg.filter("rating_count >= 100")
```

```python
movie_ratings = ratings_filtered.join(
    movies_df,
    on="movieId",
    how="inner"
)
```

```python
top_10_movies = movie_ratings \
    .orderBy("avg_rating", ascending=False) \
    .select("movieId", "title", "avg_rating", "rating_count") \
    .limit(10)
```

```python
top_10_movies.show(truncate=False)
```

```python
joined_df = ratings_df.join(
    movies_df,
    ratings_df.movieId == movies_df.movieId,
    "inner"
)
```

```python
from pyspark.sql.functions import avg, count

agg_df = joined_df.groupBy(
    movies_df.movieId,
    movies_df.title
).agg(
    avg("rating").alias("avg_rating"),
    count("rating").alias("rating_count")
)
```

```python
filtered_df = agg_df.filter("rating_count >= 100")
```

```python
top_10_movies = filtered_df \
    .orderBy("avg_rating", ascending=False) \
    .limit(10)
```

```python
top_10_movies.show(truncate=False)
```
